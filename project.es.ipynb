#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón ur de Madrid por color d cada población (debes usar el subconjunto obtenido en la pregunta 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar u código simple que transforme las columnas de coordenadas en un diccionario de python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, Marker, basemaps, Icon

map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
coordenadas_por_poblacion = {}

# Iterar sobre subconjunto_df para agrupar las coordenadas por población
for index, row in subconjunto_df.iterrows():
    # Obtener la población y las coordenadas de la fila
    poblacion = row['level5']
    latitud, longitud = row['latitude'], row['longitude']
    # Agregar las coordenadas a la lista correspondiente en el diccionario
    if poblacion in coordenadas_por_poblacion:
        coordenadas_por_poblacion[poblacion].append((latitud, longitud))
    else:
        coordenadas_por_poblacion[poblacion] = [(latitud, longitud)]

# Mostrar las primeras coordenadas para verificar
coordenadas_por_poblacion
```

```python
map = Map(center=(40.4, -3.7), zoom=10, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)

colores_por_grupo = {
    "Fuenlabrada": "blue",
    "Alcorcón": "red",
    "Leganés": "green",
    "Getafe": "purple"
}
```

```python
from ipyleaflet import Map, Marker, basemaps, DivIcon

for poblacion, coordenadas in coordenadas_por_poblacion.items():
    color = colores_por_grupo.get(poblacion, "black")  # Si la población no tiene un color asignado, se usa negro
    for latitud, longitud in coordenadas:
        marker = Marker(location=(float(latitud.replace(',', '.')), float(longitud.replace(',', '.'))),
                        draggable=False)
        
        # Crear un DivIcon personalizado con el color de la población y tamaño ajustado
        icono_personalizado = DivIcon(icon_size=(15, 15),
                                      icon_anchor=(7, 15),
                                      html=f'<div style="width: 15px; height: 15px; background-color: {color}; border-radius: 50%;"></div>')
        
        # Agregar el DivIcon como una capa adicional en el marcador
        marker.icon = icono_personalizado
        
        map.add_layer(marker)

map
```
